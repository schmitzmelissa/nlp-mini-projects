# Word Cloud Generator
## Project Purpose

This dataset contains text from emails which are marked as spam (1) or not spam (0). The purpose of this mini project is to create a word cloud to visualize key words between these two classifications.

### Initialize Project

```python
import pandas as pd
import sqlite3
import re
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# Create df from spam emails csv
df = pd.read_csv('emails.csv')

df.head()
```

### Exploratory Analysis

```python
# Checking counts IDed as spam and not spam
print("spam count: " + str(len(df.loc[df.spam==1])))
print("not spam count: " + str(len(df.loc[df.spam==0])))
print(df.shape)
df['spam'] = df['spam'].astype(int)
```

```python
# Remove duplicates and recheck shape
df = df.drop_duplicates()
df = df.reset_index(inplace = False)[['text','spam']]

print(df.shape)
```

### Data Cleaning

```python
# Clean up data so everything is lowercase with no special characters

# Create empty list for cleaned descriptions
clean_desc = []

# Goes through text line by line to...
for w in range(len(df.text)):
    desc = df['text'][w].lower() # set it to lowercase
    desc = re.sub('[^a-zA-Z]', ' ', desc) # remove punctuation
    desc = re.sub("&lt;/?.*?&gt;"," &lt;&gt; ",desc) # remove tags
    desc=re.sub("(\\d|\\W)+"," ",desc) # remove digits and special charas
    
    # Add cleaned text into list
    clean_desc.append(desc)
    
# Assign the cleaned text descriptions to the text column in df
df['text'] = clean_desc

df.head(3)
```

```python
# Create a list of stop words to omit from final word cloud
stop_words = ['is','you','your','and', 'the', 'to', 'from', 'or', 'I',
              'for', 'do', 'get', 'not', 'here', 'in', 'im', 'have',
              'on', 're', 'new', 'subject','but']
```

### Generate word cloud
First, of all the words regardless of classification.

```python
# Instantiate word cloud with chosen parameters
wordcloud = WordCloud(width = 800,
                     height = 800,
                     background_color = 'black',
                     stopwords = stop_words,
                     max_words = 1000,
                     min_font_size = 20).generate(str(df['text']))

# Plot the word cloud
fig = plt.figure(figsize = (8,8),
                facecolor = None)

plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

Second, split into spam and not spam.

```python
# Spam word cloud

# Instantiate word cloud with chosen parameters
wordcloud_spam = WordCloud(width = 800,
                     height = 800,
                     background_color = 'black',
                     stopwords = stop_words,
                     max_words = 1000,
                     min_font_size = 20).generate(str(df['text'][1]))

# Plot the word cloud
fig_spam = plt.figure(figsize = (8,8),
                facecolor = None)

plt.imshow(wordcloud_spam)
plt.axis('off')
plt.show()
```

```python
# Not spam word cloud

# Instantiate word cloud with chosen parameters
wordcloud_notspam = WordCloud(width = 800,
                     height = 800,
                     background_color = 'black',
                     stopwords = stop_words,
                     max_words = 1000,
                     min_font_size = 20).generate(str(df['text'][0]))

# Plot the word cloud
fig_notspam = plt.figure(figsize = (8,8),
                facecolor = None)

plt.imshow(wordcloud_notspam)
plt.axis('off')
plt.show()
```
